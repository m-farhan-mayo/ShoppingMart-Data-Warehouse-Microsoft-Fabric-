```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Customers Data
```

```python
df_cust = spark.read.format("csv")\
                    .option("header","true")\
                    .load("Files/ShoppingMart_Bronze_Customers/ShoppingMart_customers.csv")
display(df_cust)
```

```python
df_cust = df_cust.withColumn("CustomerName",upper(col("CustomerName")))
df_cust = df_cust.dropna(subset=["CustomerID"])
df_cust = df_cust.withColumn("Email_Domain", split("Email", '@')[1])
display(df_cust)
```

```python
df_cust.write.mode("overwrite").parquet("abfss://Farhan_Workspace@onelake.dfs.fabric.microsoft.com/ShoppingMart_Silver.Lakehouse/Files/ShoppingMart_Silver_Customers/ShoppingMart_Customers")
```

```python
#Order Data
```

```python
df_order = spark.read.format("csv")\
                     .option("header","true")\
                     .load("Files/ShoppingMart_Bronze_Orders/ShoppingMart_orders.csv")
```

```python
df_order = df_order.dropna(subset=["OrderID","OrderDate", "CustomerID", "ProductID", "TotalAmount"])
df_order = df_order.withColumn("CostPerItem", col("TotalAmount") / col("Quantity"))
df_order = df_order.withColumn("OrderDate", to_date(col("OrderDate")))
display(df_order)
```

```python
df_order.write.mode("overwrite").parquet("abfss://Farhan_Workspace@onelake.dfs.fabric.microsoft.com/ShoppingMart_Silver.Lakehouse/Files/ShoppingMart_Silver_Orders/ShoppingMart_Orders")
```
