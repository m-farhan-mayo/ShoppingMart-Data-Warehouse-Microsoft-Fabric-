```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Customers Data
```

```python
df_cust_gold = spark.read.parquet("Files/ShoppingMart_Silver_Customers/ShoppingMart_Customers")

display(df_cust_gold)
```

```python
df_cust_gold = (
    df_cust_gold
    .select("CustomerName", "Email", "SignupDate"))

display(df_cust_gold)
```

```python
df_cust_gold.write.mode("overwrite").parquet(Files/ShoppingMart_Gold_Customer/ShoppingMart_Customers)
```

```python
#Orders Data
```

```python
df_ord_gold = spark.read.parquet("Files/ShoppingMart_Silver_Orders/ShoppingMart_Orders")

display(df_ord_gold)
```

```python
df_ord_gold.write.mode("overwrite").parquet(Files/ShoppingMart_Gold_Orders/ShoppingMart_Orders)
```

```python
#Products Data
```

```python
#Aggregating Product Reviews To Calculate the Average Rate Per Product
```

```python
df_prod_gold = spark.read.parquet("Files/ShoppingMart_Silver_Products/ShoppingMart_Products")

display(df_prod_gold)
```

```python
df_prod_gold = df_prod_gold.groupBy("Category" , "Stock").count()
display(df_prod_gold)
```

```python
df_prod_gold.write.mode("overwrite").parquet(Files/ShoppingMart_Gold_Products/ShoppingMart_Products)
```

```python
#Social Medias Data
#Aggregates Unstructured Social Media Data To Track Down Trends 
```

```python
df_socialm_gold = spark.read.parquet("Files/ShoppingMart_Silver_Products/ShoppingMart_Social_Media")

display(df_socialm_gold)
```

```python
#Aggregates Web Log Data To Measure Engagment Per User On Each Page And Action
df_socialm_gold = df_socialm_gold.groupBy("content","platform","sentiment").count()

display(df_socialm_gold)
```

```python
df_socialm_gold.write.mode("overwrite").parquet(Files/ShoppingMart_Gold_Social_Media/ShoppingMart_SocialMedia)
```

```python
# Web Logs Data
```

```python
df_webl_gold = spark.read.parquet("Files/ShoppingMart_Silver_Products/ShoppingMart_Web_Logs")

display(df_webl_gold)
```

```python
#Aggregates Web Log Data To Measure Engagment Per User On Each Page And Action
df_webl_gold = df_webl_gold.groupBy("user_id","page","action").count()

display(df_webl_gold)
```

```python
df_webl_gold.write.mode("overwrite").parquet(Files/ShoppingMart_Gold_Web_Logs/ShoppingMart_Web_Logs)
```
